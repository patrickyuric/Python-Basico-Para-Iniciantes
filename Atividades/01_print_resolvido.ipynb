# Atividade: "Organize e Exiba com Print()"

Crie um pequeno programa Python para registrar e exibir 
um conjunto de informações sobre você ou um personagem fictício. 
O objetivo é praticar diferentes formas de usar o print() 
para mostrar dados, aplicar formatação e experimentar 
recursos como quebra de linha, tabulação e f-strings.

1. Crie três variáveis: nome (string), idade (inteiro), hobby (string)

```python
nome = 'Guido van Rossum'
idade = 70
hobby = 'Desenvolver Linguagens de Programação'
```

2. Exiba cada informação em uma linha separada usando três comandos print.
python

```python
print(nome)
print(idade)
print(hobby)
```

3. Mostre todas as informações em uma única linha, separando por vírgula.

```python
print(nome, ', ', idade, ', ', hobby)
```

4. Exiba os dados como uma tabela usando tabulação \t:

```python
print('Nome\tIdade\tHobby')
print(nome, idade, hobby, sep='\t')
```

5. Personalize o final da linha usando o parâmetro end, para continuar a frase:

```python
print("Guido van Rossum", end=' ')
print("é um matemático e programador", end=' ')
print("conhecido criar o Python.")
```

6. Crie uma frase formatada usando f-string:

```python
print(f'{nome} é o criador da linguagem de programação Python, está com {idade} anos e tem como hobby {hobby}.')
```
